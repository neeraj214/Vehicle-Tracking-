# 📊 UA-DETRAC Dataset Exploration

```python
import sys 
sys.path.append('..') 
from training.dataset import UATRACDataset 
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
```

```python
# Update these paths after downloading UA-DETRAC 
IMG_DIR = "../data/UA-DETRAC/images" 
ANN_DIR = "../data/UA-DETRAC/annotations" 
dataset = UATRACDataset(IMG_DIR, ANN_DIR) 
print(f"Total samples: {len(dataset)}")
```

```python
# Visualize one frame with bounding boxes 
import torchvision.transforms.functional as F 
img, boxes, track_ids = dataset[0] 
fig, ax = plt.subplots(1, figsize=(12, 6)) 
ax.imshow(F.to_pil_image(img)) 
for box, tid in zip(boxes, track_ids): 
    rect = patches.Rectangle( 
        (box[0], box[1]), box[2], box[3], 
        linewidth=2, edgecolor='lime', facecolor='none' 
    ) 
    ax.add_patch(rect) 
    ax.text(box[0], box[1]-5, f"ID:{tid}", 
            color='lime', fontsize=8, fontweight='bold') 
plt.title("UA-DETRAC Frame with Ground Truth Tracks") 
plt.axis('off') 
plt.show()
```
